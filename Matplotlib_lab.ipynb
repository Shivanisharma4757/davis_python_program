```
import matplotlib.pyplot as plt

# Input: x and y values
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot")

# Show plot
plt.show()
```

```
import matplotlib.pyplot as plt

# Input: Product names and their sales
products = ["Product A", "Product B", "Product C", "Product D", "Product E"]
sales = [100, 150, 80, 200, 120]

# Create bar chart
plt.bar(products, sales)

# Add labels and title
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales Bar Chart")

# Show plot
plt.show()
```

```
import matplotlib.pyplot as plt

# Input: Category-wise data
categories = ["Food", "Transport", "Rent", "Shopping", "Others"]
values = [300, 100, 500, 200, 150]

# Create pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%')

# Add title
plt.title("Expense Distribution")

# Show plot
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Input: Random dataset
data = np.random.randn(1000)  # 1000 random values (normal distribution)

# Create histogram
plt.hist(data, bins=20)

# Add labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Dataset")

# Show plot
plt.show()
```

```
import matplotlib.pyplot as plt

# Input: Two datasets
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

# Plot both datasets
plt.plot(x, y1, label="Dataset 1")
plt.plot(x, y2, label="Dataset 2")

# Add labels, title, and legend
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Comparison of Two Datasets")
plt.legend()

# Show plot
plt.show()
```

```
import matplotlib.pyplot as plt

# Input: Data points
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

# Find maximum point
max_y = max(y)
max_index = y.index(max_y)
max_x = x[max_index]

# Create scatter plot
plt.scatter(x, y)

# Highlight maximum point
plt.scatter(max_x, max_y, s=100)  # larger size to highlight

# Add labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot with Maximum Point Highlighted")

# Annotate the max point
plt.text(max_x, max_y, f"({max_x}, {max_y})")

# Show plot
plt.show()
```

```
import matplotlib.pyplot as plt

# Input: Monthly revenue data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
revenue = [200, 250, 300, 280, 350, 400]

# Create line plot
plt.plot(months, revenue)

# Customize plot
plt.title("Monthly Revenue")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.grid(True)

# Show plot
plt.show()
```
